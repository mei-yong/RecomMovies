# Helper functions to QA the Neo4j graph db creation

```python
def check_dupes(label, uid):
    # Get number of all nodes of a particular label
    q_all_nodes = "MATCH (n:" + label + ") RETURN COUNT(n)"
    all_nodes = graph.run(q_all_nodes).evaluate()
    
    q_distinct_nodes = "MATCH (n:" + label + ") RETURN COUNT(DISTINCT n." + uid + ")"
    distinct_nodes = graph.run(q_distinct_nodes).evaluate()
    
    result = {'all_nodes': all_nodes, 'distinct_nodes': distinct_nodes}
    print(result)
```

# Build the base graph db

```python
# Create uniqueness constraints & indexes
graph.run("CREATE CONSTRAINT ON (m:Movie) ASSERT m.id IS UNIQUE")
graph.run("CREATE CONSTRAINT ON (u:User) ASSERT u.id IS UNIQUE")
graph.run("CREATE CONSTRAINT ON (a:Artist) ASSERT a.name IS UNIQUE")
graph.run("CREATE CONSTRAINT ON (g:Genre) ASSERT g.name IS UNIQUE")
```

```python
# Create movie nodes
query = '''
USING PERIODIC COMMIT 500
LOAD CSV WITH HEADERS
FROM "file:///movies_imdb.csv" AS row
MERGE (m:Movie { id: toInteger(row.netflix_id) })
ON CREATE SET
    m.imdb = row.imdb_id,
    m.title = row.title,
    m.year = toInteger(row.year)
'''
graph.run(query)
```

```python
# QA movie nodes
check_dupes(label='Movie', uid='id')
```

```python
# Create user nodes
query = '''
USING PERIODIC COMMIT 500
LOAD CSV WITH HEADERS
FROM "file:///users.csv" AS row
MERGE (u:User {id:toInteger(row.user_id)})
ON CREATE SET u.name = trim(row.name)
'''
graph.run(query)
```

```python
# QA user nodes
check_dupes(label='User', uid='id')
```

```python
# Create rating relationships
query = '''
USING PERIODIC COMMIT 500
LOAD CSV WITH HEADERS
FROM "file:///ratings.csv" AS row
WITH row, SPLIT(row.rating_date, "-") AS date
MATCH (u:User {id: toInteger(row.user_id)})
MATCH (m:Movie {id: toInteger(row.movie_id)})
MERGE (u)-[r:RATED]->(m)
ON CREATE SET
    r.rating = toInteger(row.rating),
    r.year = toInteger(date[0]),
    r.month = toInteger(date[1]),
    r.day = toInteger(date[2])
'''
graph.run(query)
```

```python
# QA rating relationships
graph.run("MATCH ()-[r:RATED]-() RETURN COUNT(DISTINCT r)").evaluate()
```

```python
# Add average rating & total ratings to movie nodes
query = '''
MATCH (m:Movie)<-[r:RATED]-()
WITH m, 
    AVG(r.rating) as average_rating, 
    COUNT(r.rating) as total_ratings
SET m.avg_rating = round(100* average_rating) / 100,
    m.total_ratings = total_ratings
'''
graph.run(query)
```

```python
# Get top 10 highly rated movies
query = '''
MATCH (m:Movie)
RETURN m.title as movie, m.avg_rating as rating
ORDER BY m.avg_rating DESC LIMIT 10
'''
result = graph.run(query)

for row in result:
    print(row)
```

# Content Filtering
We're going to go with Oswaldo as an example case - because I like his name

```python
# Find other movies with similar genres
query = '''
MATCH (user:User)-[:RATED]->(m:Movie),
    (m)-[:HAS_GENRE]->(g:Genre)<-[:HAS_GENRE]-(otherMovie:Movie)
WHERE user.name CONTAINS "Oswaldo"
    AND NOT (user)-[:RATED]->(otherMovie)

WITH otherMovie, [g.name, COUNT(*)] AS scores

RETURN
    otherMovie.title as recommendation,
    COLLECT(scores) AS scoreComponents,
    REDUCE (s=0, x in COLLECT(scores) | s+x[1]) AS score

ORDER BY score DESC
LIMIT 10
'''
results = graph.run(query)

df = pd.DataFrame(results, columns=['recommendation','scoreComponents','score'])
df.head(10)
```

```python
# Find other movies of similar genres with high ratings
query = '''
MATCH (user:User)-[:RATED]->(m:Movie)-[:HAS_GENRE]->(g:Genre)<-[:HAS_GENRE]-(otherMovie:Movie)
WHERE user.name CONTAINS "Oswaldo"
    AND NOT (user)-[:RATED]->(otherMovie)
    AND otherMovie.avg_rating > 3
RETURN 
    COLLECT(m.title) AS watchedMovies,
    otherMovie.title AS recommendation,
    otherMovie.avg_rating AS averageRating,
    COUNT(DISTINCT g) AS genresInCommon,
    COLLECT(DISTINCT g.name) AS genres
ORDER BY genresInCommon DESC, averageRating
LIMIT 10
'''
results = graph.run(query)

df = pd.DataFrame(results, columns=['watchedMovies','recommendation','averageRating','genresInCommon','genres'])
df.head(10)
```

```python
# Find other movies with similar genres, actors, directors combined into a weighted score
query = '''
MATCH (user:User)-[:RATED]->(m:Movie),
    (m)-[:HAS_GENRE]->(g:Genre)<-[:HAS_GENRE]-(otherMovie:Movie)
WHERE user.name CONTAINS "Oswaldo"
    AND NOT (user)-[:RATED]->(otherMovie)

WITH m, otherMovie, COUNT(*) AS genres

OPTIONAL MATCH (m)<-[:ACTED_IN]-(a:Artist)-[:ACTED_IN]->(otherMovie)
WITH m, otherMovie, genres, COUNT(*) AS actors

OPTIONAL MATCH (m)<-[:DIRECTED]-(d:Artist)-[:DIRECTED]->(otherMovie)
WITH m, otherMovie, genres, actors, COUNT(*) AS directors

RETURN
    COLLECT(m.title) AS watchedMovies,
    otherMovie.title as recommendation,
    (5*genres)+(3*actors)+(4*directors) AS weightedScore

ORDER BY weightedScore DESC
LIMIT 10
'''
results = graph.run(query)

df = pd.DataFrame(results, columns=['watchedMovies','recommendation','weightedScore'])
df.head(10)
```

```python
# Find other movies with similar genres - but this time using jaccard scoring
query = '''
MATCH (user:User)-[:RATED]->(m:Movie),
    (m)-[:HAS_GENRE]->(g:Genre)<-[:HAS_GENRE]-(otherMovie:Movie)
WHERE user.name CONTAINS "Oswaldo"
    AND NOT (user)-[:RATED]->(otherMovie)
    
WITH m, otherMovie, COUNT(g) AS intersection
MATCH (m)-[:HAS_GENRE]->(movieGenre:Genre)
WITH m, otherMovie, intersection, COLLECT(movieGenre.name) AS s1
MATCH (otherMovie)-[:HAS_GENRE]->(otherGenre:Genre)
WITH m, otherMovie, intersection, s1, COLLECT(otherGenre.name) AS s2
WITH m, otherMovie, intersection, s1, s2, s1+filter(x IN s2 WHERE NOT x IN s1) AS union

RETURN m.title,
    otherMovie.title,
    s1,
    s2,
    ((1.0*intersection)/SIZE(union)) AS jaccard
ORDER BY jaccard DESC
LIMIT 10
'''
results = graph.run(query)

df = pd.DataFrame(results, columns=['watchedMovie','recommendation','watchedGenres','recommendedGenres','jaccardScore'])
df.head(10)
```

```python
# Find other movies with similar genres, actors, directors - but this time using jaccard scoring
query = '''
MATCH (user:User)-[:RATED]->(m:Movie),
    (m)-[:HAS_GENRE|:ACTED_IN|:DIRECTED]-(tag)-[:HAS_GENRE|:ACTED_IN|:DIRECTED]-(otherMovie:Movie)
WHERE user.name CONTAINS "Oswaldo"
    AND NOT (user)-[:RATED]->(otherMovie)
    
WITH m, otherMovie, COUNT(tag) AS intersection
MATCH (m)-[:HAS_GENRE|:ACTED_IN|:DIRECTED]-(movieTag)
WITH m, otherMovie, intersection, COLLECT(movieTag.name) AS s1
MATCH (otherMovie)-[:HAS_GENRE|:ACTED_IN|:DIRECTED]-(otherTag)
WITH m, otherMovie, intersection, s1, COLLECT(otherTag.name) AS s2
WITH m, otherMovie, intersection, s1, s2, s1+filter(x IN s2 WHERE NOT x IN s1) AS union

RETURN m.title,
    otherMovie.title,
    s1,
    s2,
    ((1.0*intersection)/SIZE(union)) AS jaccard
ORDER BY jaccard DESC
LIMIT 10
'''
results = graph.run(query)

df = pd.DataFrame(results, columns=['watchedMovie','recommendation','watchedTags','recommendedTags','jaccardScore'])
df.head(10)
```

```python
# What genres does the user like more than average? Find other movies with said genre(s)
query = '''
MATCH (user:User)-[r:RATED]->(m:Movie) WHERE user.name CONTAINS "Oswaldo"
WITH user, AVG(r.rating) AS mean

MATCH (user:User)-[r:RATED]->(m:Movie)-[:HAS_GENRE]->(g:Genre)
WHERE r.rating > mean

WITH user, g, COUNT(*) AS score

MATCH (g)<-[:HAS_GENRE]-(otherMovie:Movie)
WHERE NOT (user)-[:RATED]->(otherMovie)

RETURN otherMovie.title AS recommendation,
    COLLECT(DISTINCT g.name) AS genres,
    SUM(score) AS score
ORDER BY score DESC
LIMIT 10
'''
results = graph.run(query)

df = pd.DataFrame(results, columns=['recommendation','genres','score'])
df.head(10)
```

# Collaborative Filtering
We're going to go with Oswaldo again

```python
# Find most similar users using Cosine similarity - takes into account ratings of the individuals (not just which movies they have in common)
query = '''
MATCH (p1:User)-[x:RATED]-(m:Movie)<-[y:RATED]-(p2:User) WHERE p1.name CONTAINS "Oswaldo"
WITH COUNT(m) AS numberMovies,
    SUM(x.rating * y.rating) AS xyDotProduct,
    SQRT(REDUCE(xDot=0.0, a IN COLLECT(x.rating) | xDot + a^2)) AS xLength,
    SQRT(REDUCE(yDot=0.0, b IN COLLECT(y.rating) | yDot + b^2)) AS yLength,
    p1,
    p2
    WHERE numberMovies > 10

RETURN p1.name,
    p2.name,
    xyDotProduct/ (xLength * yLength) AS cosine
ORDER BY cosine DESC
LIMIT 10
    
'''
results = graph.run(query)

df = pd.DataFrame(results, columns=['person1','person2','cosSimScore'])
df.head(10)
```

```python
# Find most similar users using Cosine similarity - takes into account ratings of the individuals (not just which movies they have in common)
# Using the Neo4j Enterprise algorithm
query = '''
MATCH (p1:User)-[x:RATED]-(m:Movie)<-[y:RATED]-(p2:User)
WHERE p1.name CONTAINS "Oswaldo"
    AND p1 <> p2
WITH p1, p2, COLLECT(x.rating) AS p1Ratings, COLLECT(y.rating) AS p2Ratings
    WHERE SIZE(p1Ratings) > 10

RETURN p1.name,
    p2.name,
    algo.similarity.cosine(p1Ratings, p2Ratings) AS cosine
ORDER BY cosine DESC
LIMIT 10
'''
results = graph.run(query)

df = pd.DataFrame(results, columns=['person1','person2','cosSimScore'])
df.head(10)
```

```python
# Find most similar users using Pearson similarity - takes into account ratings of the individuals PLUS that different users have different average ratings
query = '''
MATCH (p1:User)-[r:RATED]->(m:Movie) WHERE p1.name CONTAINS "Oswaldo"
WITH p1, AVG(r.rating) AS p1_mean

MATCH (p1)-[x:RATED]->(m:Movie)<-[y:RATED]-(p2)
WITH p1, p1_mean, p2, COLLECT({x:x, y:y}) AS ratings WHERE SIZE(ratings) > 10

MATCH (p2)-[r:RATED]->(m:Movie)
WITH p1, p1_mean, p2, AVG(r.rating) AS p2_mean, ratings 

UNWIND ratings AS r

WITH SUM( (r.x.rating - p1_mean) * (r.y.rating - p2_mean) ) AS nom,
    SQRT( SUM( (r.x.rating - p1_mean)^2) * SUM( (r.y.rating - p2_mean)^2) ) AS denom,
    p1,
    p2 
    WHERE denom <> 0
    
RETURN p1.name,
    p2.name,
    nom/denom AS pearson
ORDER BY pearson DESC
LIMIT 10
'''
results = graph.run(query)

df = pd.DataFrame(results, columns=['person1','person2','pearsonScore'])
df.head(10)
```

```python
# Find most similar users using Pearson similarity - takes into account ratings of the individuals PLUS that different users have different average ratings
# Using the Neo4j Enterprise algorithm
query = '''
MATCH (p1:User)-[x:RATED]-(m:Movie) WHERE p1.name CONTAINS "Oswaldo"
WITH p1, algo.similarity.asVector(m, x.rating) AS p1Vector
MATCH (p2:User)-[y:RATED]->(m:Movie) WHERE p2 <> p1
WITH p1, p2, p1Vector, algo.similarity.asVector(m, y.rating) AS p2Vector
WHERE SIZE(apoc.coll.intersection( [v in p1Vector | v.category], [v in p2Vector | v.category])) > 10

RETURN p1.name,
    p2.name,
    algo.similarity.pearson(p1Vector, p2Vector, {vectorType: "maps"}) AS pearson
ORDER BY pearson DESC
LIMIT 10
'''
results = graph.run(query)

df = pd.DataFrame(results, columns=['person1','person2','pearsonScore'])
df.head(10)
```

```python
# kNN - Find the top 10 most similar users to our target user using Peason similarity and then find other movies that said users rated highly
query = '''
MATCH (p1:User)-[r:RATED]->(m:Movie) WHERE p1.name CONTAINS "Oswaldo"
WITH p1, AVG(r.rating) AS p1_mean

MATCH (p1)-[x:RATED]->(m:Movie)<-[y:RATED]-(p2)
WITH p1, p1_mean, p2, COLLECT({x:x, y:y}) AS ratings WHERE SIZE(ratings) > 10

MATCH (p2)-[r:RATED]->(m:Movie)
WITH p1, p1_mean, p2, AVG(r.rating) AS p2_mean, ratings 

UNWIND ratings AS r

WITH SUM( (r.x.rating - p1_mean) * (r.y.rating - p2_mean) ) AS nom,
    SQRT( SUM( (r.x.rating - p1_mean)^2) * SUM( (r.y.rating - p2_mean)^2) ) AS denom,
    p1,
    p2 
    WHERE denom <> 0

WITH p1, p2, nom/denom AS pearson
ORDER BY pearson DESC 
LIMIT 10

MATCH (p2)-[r:RATED]->(m:Movie) WHERE NOT EXISTS((p1)-[:RATED]->(m))

RETURN m.title, SUM(pearson * r.rating) AS score
ORDER BY score DESC
LIMIT 10
'''
results = graph.run(query)

df = pd.DataFrame(results, columns=['recommendation','score'])
df.head(10)
```
